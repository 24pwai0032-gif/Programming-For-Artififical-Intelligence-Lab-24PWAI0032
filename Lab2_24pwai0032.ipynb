## Task 02 — Mini Calculator with Validation

```python
# Collect the two numbers — float() as an input
num1 = float(input("Enter the first number:  "))
num2 = float(input("Enter the second number: "))

# Asking which operation to perform
op = input("Choose an operation  ( + - * / ** % ):  ").strip()

# List of all operators
valid_ops = ["+", "-", "*", "/", "**", "%"]

# operator must be one i mentioned/recognized
if op not in valid_ops:
    print("Invalid operator!")
elif op == "/" and num2 == 0:
    print("Cannot divide by zero!")
else:
    result = eval(f"{num1} {op} {num2}")

    if op in ["/", "**"]:
        print(f"\n  {num1} {op} {num2} = {result:.2f}")
    else:
        print(f"\n  {num1} {op} {num2} = {result}")
```

## Task 03 — Challenge A: Even / Odd & Multiple Checker

```python
# Get the range from the user
start = int(input("Enter the starting number: "))
end   = int(input("Enter the ending number:   "))

print(f"\nChecking numbers from {start} to {end}...\n")
print(f"{'Number':<10} {'Result'}")
print("-" * 35)

current = start   
# using a while loop to check each number in the range
while current <= end:
    if current % 7 == 0:
        current += 1
        continue          # jump to next iteration

    # Now figure out which label fits this number
    is_even       = (current % 2 == 0)
    is_multiple_5 = (current % 5 == 0)

    if is_even and is_multiple_5:
        label = " Even & Multiple of 5!!"
    elif is_multiple_5:
        label = " Multiple of 5!"
    elif is_even:
        label = "Even"
    else:
        label = "Odd"

    print(f"{current:<10} {label}")

    current += 1

print("\nDone!")
```

##  Task 03 — Challenge B: Password Strength Checker

```python
password = input("Enter a password to check its strength: ")

print() 

# Check 1: length
# A password under 6 characters is too weak
if len(password) < 6:
    print("Too short!  A good password needs at least 6 characters.")
else:
    # Check 2: must contain at least one digit
    has_digit     = any(ch.isdigit() for ch in password)

    # Check 3: must contain at least one uppercase letter
    has_upper     = any(ch.isupper() for ch in password)

    if not has_digit:
        print("Add at least one number")

    elif not has_upper:
        print("Add at least one capital letter.")

    # All checks passed!
    else:
        print("  Strong password good choice!")
```
