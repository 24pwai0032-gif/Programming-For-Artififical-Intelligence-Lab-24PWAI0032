### Task 1 – Even or Odd

```python
number = int(input("Enter a number: "))

if number % 2 == 0:
    print(f"{number} is Even")
else:
    print(f"{number} is Odd")
```

### Task 2 – Grade System

```python
marks = float(input("Enter your marks (0-100): "))

if marks >= 80:
    grade = "A"
elif marks >= 60:
    grade = "B"
elif marks >= 50:
    grade = "C"
else:
    grade = "Fail"

print(f"Your grade is: {grade}")
```

### Task 3 – Voting Eligibility

###  Task 7 – Password Checker (While Loop + Condition)

```python
CORRECT_PASSWORD = "admin1234"

while True:
    entered = input("Enter the password: ")
    if entered == CORRECT_PASSWORD:
        print(" Access Granted! Welcome.")
        break
    else:
        print(" Wrong password. Please try again.")
```

###  Task 8 – Function: `average(a, b, c)`
Write a function that takes three numbers and **returns** their average.

```python
def average(a, b, c):
    """
    Calculate and return the average of three numbers.
    Parameters: a, b, c – numeric values
    Returns: float – arithmetic mean of a, b, c
    """
    return (a + b + c) / 3


# Test the function 
x = float(input("Enter first value: "))
y = float(input("Enter second value: "))
z = float(input("Enter third value: "))

result = average(x, y, z)
print(f"\nAverage of {x}, {y}, {z}  →  {result:.2f}")

# Additional test cases
print("\nBuilt-in test cases")
print("average(10, 20, 30) =", average(10, 20, 30))   # Expected: 20.0
print("average(5,  5,  5 ) =", average(5,  5,  5))    # Expected:  5.0
print("average(0, 100, 50) =", average(0, 100, 50))   # Expected: 50.0
```
